```python
import datetime
import pandas as pd
```

```python
start = datetime.datetime(2016, 1, 1)
stop = datetime.date.today()
```

```python
stop
```

```python
apple = pdr.get_data_google('AAPL', start, stop)
microsoft = pdr.get_data_google('MSFT', start, stop)
google = pdr.get_data_google('GOOG', start, stop)
nvidia = pdr.get_data_google("NVDA", start, stop)
stocks = pd.DataFrame({"AAPL": apple['Close'],
                       "MSFT": microsoft['Close'],
                       "GOOG": google['Close'],
                       "NVDA": nvidia['Close']})
stocks.head()
```

```python
%matplotlib inline
```

```python
stocks.plot(secondary_y = ["AAPL", "MSFT"], grid=True)
```

```python
stock_return = stocks.apply(lambda x: x/x[0])
stock_return.head()
```

```python
stock_return.plot(grid = True)
```

```python
stock_return[['AAPL', 'GOOG', 'MSFT']].plot(grid = True)
```

```python
import numpy as np
```

```python
stock_change = stocks.apply(lambda x: np.log(x) - np.log(x.shift(1)))
stock_change.head()
```

```python
stock_change.plot(grid=True).axhline(y = 0, color = 'black', lw=1)
```

```python
def plot_ohlc_test(stock, mov_avg):
    '''
    '''
    
    
    stock_data = pdr.get_data_google(stock)
    
    stock_data[str(mov_avg)] = np.round(stock_data["Close"].rolling(window = mov_avg, center = False).mean(), 2)
    
    
    # return stock_data
    # stock_data['Date'] = stock_data.index
    time_stamp = mdates.date2num(stock_data.index.to_pydatetime())
    # drop the date index from the dataframe
    stock_data.reset_index(inplace=True)
    stock_data['Date'] = time_stamp
    
    fig = plt.figure()
    ax1 = plt.subplot2grid((1,1), (0,0))
    
    
    # x = 0
    # y = len(stock_data)
    
    # while x < y:
     #   append_me = 
    
    #ax1.xaxis.label.set_color('c')
    #ax1.yaxis.label.set_color('r')
    #ax1.set_yticks([0, 25, 50, 75])
    
    candlestick_ohlc(ax1, stock_data.values, width=0.4, colorup='#77d879', colordown='#db3f3f')
    
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
        
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    
    
    if mov_avg != None:
        ax1.plot(stock_data['Date'], stock_data[str(mov_avg)])
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(stock)
    # plt.legend()
    

        
    plt.show()
```

```python
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
from matplotlib.finance import candlestick_ohlc
from matplotlib import style
```

```python
plot_ohlc_test('NVDA', mov_avg=300)
```
